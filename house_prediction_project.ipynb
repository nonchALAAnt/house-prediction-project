```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = pd.read_csv("houses.csv")

x = np.array( file[[ 'bedrooms' , 'sqft_living' , 'floors' , 'condition' ]].values )
# we can add more parameters to make the prediction more acurate but I believe those 4 are the most important elements

y = np.array( file['price'])

mask = ~np.isnan(x).any(axis=1)

x = x[mask]
y = y[mask]

mean = np.mean(x, axis=0)
std = np.std(x, axis=0)

std[std == 0] = 1 

x = (x - mean) / std

m = len(y)
n = len(x[0])

w = np.zeros(n)
b = 0.

costs = []

def jcost(w, b):
    predictions = x @ w + b
    return np.sum((predictions - y)**2) / (2*m)

# this function calculates the cost of the current funciton f(w,b)
# the cost function is the sum of the squared difference between our prediction funtion and the actual data devided by 2m
# the less the cost function is the more acurate the prediction is 

def gradient(w, b):
    predictions = x @ w + b
    errors = predictions - y
    dw = (x.T @ errors) / m
    db = np.sum(errors) / m
    return dw, db

alpha = 0.001
# alpha is the learning rate which controls the speed of the gradient descent funtion 

for i in range(100000):
    dw, db = gradient(w, b)
    w = w - alpha * dw
    b = b - alpha * db

    if i % 10000 == 0:
        cost = jcost(w,b)
        costs.append(cost)
        print(f"Iteration {i}: Cost = {cost}")
# the gradient descent function's job is to minimize the cost as much as poosible to acheive the best predictions possible

# this grid shows how the cost function is getting minimized 

plt.plot([i * 10000 for i in range(len(costs))], costs)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Training Cost Over Iterations")
plt.savefig("cost_curve.png")
plt.show()

bedrooms = float( input('enter the number of bedrooms') )
sqft_living =float( input('enter the size of house in squarefoots') )
floor = float( input('enter the number of floors') )
condition = float( input('enter the condition of the house from 0 to 10') )
values = np.array([bedrooms , sqft_living , floor , condition ])
values = (values - mean) / std
print('the expected price of the house is : ' , int( ( values @ w + b) * 1000) , 'USD')

```
